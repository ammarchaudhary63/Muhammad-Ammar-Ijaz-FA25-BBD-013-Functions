```python
import datetime

def greet_with_time(name):
    current_time = datetime.datetime.now().strftime("%H:%M:%S")
    print(f"Hello {name}, the current time is {current_time}.")
greet_with_time("AMMAR")
```

```python
def calculate_average(a, b, c):
    return (a + b + c) / 3
result=calculate_average(14,17,25)
print(result)
```

```python
def calculate_discount(price, discount=10):
    return price - (price * (discount / 100))
result=calculate_discount(1000)
print(result)
```

```python
def print_info(name, age, city):
    print(f"Name: {name}, Age: {age}, City: {city}")
result=print_info(name="AMMAR",age=18,city="LAHORE")
```

```python
def find_max(*num):
    return max(num)
result=find_max(100,-1,499)
print(result)
```
